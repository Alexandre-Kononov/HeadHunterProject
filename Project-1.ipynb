1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#Функция-преобразование данных к заданному виду столбца "Образование"
def cat_precision (x):
    x = x.lower()
    if x.split()[1] == 'образование': #Определение категорий "высшее" и "среднее"
        cat = x.split()[0]
    else:
        cat = x.split()[0]+ (' ') + x.split()[1] #Выделение категорий из 2-х слов по остаточному принципу  
    return cat

#Выделение столбца "Образование" и проверка результата:
hh_source_db['Образование'] = hh_source_db['Образование и ВУЗ'].apply(cat_precision)
display(hh_source_db['Образование'].unique())

#Ответ на проверочный вопрос "Задание 3.1":
display(hh_source_db['Образование'].value_counts())

#Удаление столбца "Образование и ВУЗ" и сохранение изменений в новую таблицу:
hh_modified_db = hh_source_db.drop(labels='Образование и ВУЗ', axis=1, inplace=False)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#Преобразование "Не указано" в NaN:
hh_modified_db['Опыт работы'] = hh_modified_db['Опыт работы'].apply(lambda x: np.nan if x == 'Не указано' else x)

def exp_months(s):
    
    #Списки слов для определения временного промежутка:
    years = ("год", "года", "лет")
    months = ("месяц", "месяца", "месяцев")

    if s is np.nan: #Проверка на наличие пропуска:
        return None
    
    #Определение временного промежутка и соответствующего количества месяцев:
    text = s.split()

    if text[3] in years:
        answer = int(text[2]) * 12
        if text[5] in months:
            answer += int(text[4])
    elif text[3] in months:
        answer = int(text[2])
    return answer

hh_modified_db["Опыт работы (месяц)"] = hh_modified_db['Опыт работы'].apply(exp_months)

#Удаление столбца "Опыт работы" и сохранение изменений:
hh_modified_db.drop(labels='Опыт работы', axis=1, inplace=True)
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    * … , готов к командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#Функция выделения элементов для заполнения признака "Город":
def city (x):
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 
                      'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    text =  x.split()
    if text[0] in million_cities:
        return 'город-миллионник'
    elif text[0] == "Москва":
        return 'Москва'
    elif text[0] == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    else:
        return 'другие'
    
#Функция выделения элементов для заполнения признака "Готовность к переезду":  
def move(x):
    full_text = x.split(',')
    
    for a in range(len(full_text)): #Проход по всем блокам текста из-за возможных изменений структуры ответа:  
        text_elem = full_text[a].split()
        
        if "переезду" in text_elem: #Поиск ключевого слова
            if 'не' in text_elem: #Проверка на отрицание
                return False
            else:
                return True  
            
        elif "переехать" in text_elem: #Поиск ключевого слова
            if 'не' in text_elem: #Проверка на отрицание
                return False
            else:
                return True  
            
        else: #Увеличение счётчика
            a += 1  

#Функция выделения элементов для заполнения признака "Готовность к командировкам":
def business_trip(x):
    full_text = x.split(',')
    found = False  # флажок на случай, если информация "потерялась"
    
    for a in range(len(full_text)):  
        text_elem = full_text[a].split()
        
        if "командировкам" in text_elem:
            if 'не' in text_elem:
                trip = False
            else:
                trip = True  
            found = True  # флажок срабатывает после нахождения слова
            break  # выход из цикла после нахождения искомого слова
    
    if not found:  # если искомое слово не найдено, считаем что соискатель не готов к командировкам
        trip = False 
        
    return trip

#Выделение и наполнение признаков "Город", "Готовность к переезду", "Готовность к командировкам":              
hh_modified_db['Город']= hh_modified_db['Город, переезд, командировки'].apply(city)
hh_modified_db['Готовность к переезду']= hh_modified_db['Город, переезд, командировки'].apply(move)
hh_modified_db['Готовность к командировкам']= hh_modified_db['Город, переезд, командировки'].apply(business_trip)

#Удаление столбца 'Город, переезд, командировки':
hh_modified_db.drop(labels='Город, переезд, командировки', axis=1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
element_list = [] #Поиск всех уникальных вариантов в столбце "Занятость":
column = hh_modified_db['Занятость']

for i in range(len(column)):#Просмотр всех строк столбца и добавление новых уникальных элементов в список
    text = column.iloc[i]
    words = [word.strip() for word in text.split(',')]
    for word in words:
        if word not in element_list:  
            element_list.append(word)

element_list1 = [] #Поиск всех уникальных вариантов в столбце "График":
column = hh_modified_db['График']

for i in range(len(column)):#Просмотр всех строк столбца и добавление новых уникальных элементов в список
    text = column.iloc[i]
    words = [word.strip() for word in text.split(',')]
    for word in words:
        if word not in element_list1:  
            element_list1.append(word)

#Создание новых столбцов из отобранных уникальных элементов для "One Hot Encoding" и заполнение созданных столбцов:
for col in element_list:
    hh_modified_db[col] = False
    hh_modified_db[col] = hh_modified_db['Занятость'].apply(lambda x: True if col in x.split(', ') else False)
    
for col in element_list1:
    hh_modified_db[col] = False
    hh_modified_db[col] = hh_modified_db['График'].apply(lambda x: True if col in x.split(', ') else False)
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#Выделение даты в признаке "Обновление резюме" и её приведение к формату datetime:
hh_modified_db['Обновление резюме'] = hh_modified_db['Обновление резюме'].apply(lambda x: x.split()[0])
hh_modified_db['Обновление резюме'] = pd.to_datetime(hh_modified_db['Обновление резюме'], dayfirst=True)
hh_modified_db['Обновление резюме'].head(2)

#Чтение таблицы с курсами и приведение признака "date" к формату datetime:
x_rate_db = pd.read_csv('data\ExchangeRates.csv')
x_rate_db['date'] = pd.to_datetime(x_rate_db['date'], dayfirst=True)

hh_modified_db['currency'] = hh_modified_db['ЗП'].apply(lambda x: x.split()[1]) #Выделение наименования валюты в признак "currency"
hh_modified_db['ЗП'] = hh_modified_db['ЗП'].apply(lambda x: int(x.split()[0])) #Выделение суммы желаемой зарплаты

#Функция-преобразователь для гармонизации наименования валют в таблицах:
def converter (x):
    convert_dict = {'грн.':'UAH','USD':'USD','EUR':'EUR','бел.руб.':'BYN',
                    'KGS':'KGS','сум':'UZS','AZN':'AZN','KZT':'KZT','руб.':'RUR'}
    return convert_dict[x]

hh_modified_db['currency'] = hh_modified_db['currency'].apply(converter)

#Совмещение таблицы с данными HeadHunter и таблицы с курсом валют: 
hh_modified_db = pd.merge(hh_modified_db, x_rate_db, 
                        left_on=['Обновление резюме', 'currency'], 
                        right_on=['date', 'currency'], how='left')

#Заполнение отсутствующих данных для валюты "рубль":
hh_modified_db['close'] = hh_modified_db.apply(lambda x: 1 if x['currency'] == 'RUR'
                                           else x['close'], axis=1)
hh_modified_db['proportion'].fillna(1, inplace = True)

#Создание нового признака желаемой зарплаты в рублях - "ЗП (руб)":
hh_modified_db['ЗП (руб)'] = (hh_modified_db['ЗП']*hh_modified_db['close'])/hh_modified_db['proportion']

#Очищение объединённой таблицы от лишних привнесённых признаков:
hh_modified_db.drop(labels=['per', 'date', 'time', 'vol', 'ЗП', 'close', 'proportion'], axis=1, inplace=True)
display(hh_modified_db.head())
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
#Функция определения выбросов методом z-отклонений для логарифмического преобразования данных:
def outliers_z_score_mod(data, feature, log_scale=False, left = 3, right = 3):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

#Построение гистограммы с логарифмическим масштабом:
fig, ax = plt.subplots()
ax.hist(np.log(hh_modified_db['Возраст']), bins=20)
ax.set_xlabel('Логарифм возраста')
ax.set_ylabel('Количество резюме')
ax.set_title('Распределение возраста в логарифмическом масштабе')

#Вычисление среднего и границ интервала трех сигм:
age_mean_log = np.log(hh_modified_db['Возраст']).mean()
age_std_log = np.log(hh_modified_db['Возраст']).std()
age_upper_bound = age_mean_log + 3 * age_std_log
age_lower_bound = age_mean_log - 3 * age_std_log

#Добавление линий на графике:
ax.axvline(age_mean_log, color='r', lw=1, ls='-', label = 'Медиана')
ax.axvline(age_upper_bound, color='g', lw=1, ls=':', label='Границы 3 сигм')
ax.axvline(age_lower_bound, color='g', lw=1, ls=':')
ax.legend()

plt.show()


outliers, cleaned = outliers_z_score_mod(hh_modified_db, 'Возраст', left=3, right=4, log_scale=True)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
display(outliers)
```

В какую сторону асимметрично логарифмическое распределение?

Если рассматривать симметрию графика с точки зрения распределения событий относительно медианы, то график смещён вправо.

Если рассматривать симметрию графика с точки зрения численного  распределения событий, то, судя по высоте столбцов, симметрия нарушена слева, где распределение демонстрирует значительное количество событий во второй и третьей третях левой стороны.
